```python
# remove user_handles
data["user_handles"] = data["movie_title"].apply(nfx.remove_userhandles)
```

```python
# extract hashtags
data["hash_tags"] = data["movie_title"].apply(nfx.extract_hashtags)
```

```python
# remove hashtags
data["hash_tags"] = data["movie_title"].apply(nfx.remove_hashtags)
```

```python
# remove special_chars
data["special_chars"] = data["movie_title"].apply(nfx.remove_special_characters)
```

```python
# remove multiple_spaces
data["multiple_spaces"] = data["movie_title"].apply(nfx.remove_multiple_spaces)
```
