```python
data.columns.difference(["movie_title"])  #select columns except movie_title
```

```python
data.loc[:, data.columns.isin(["DDuration"])] #select columns name
```

```python
data.loc[:, ~data.columns.isin(["DDuration"])].columns #select column names except one using negation of isin
```

```python
data.columns.str.startswith("movie") #select column names that begin with a word or char
```

```python
data.filter(like = "ration").columns  #select column names that begin with this word or char
```

```python
data.columns.str.endswith("tion") #select column names that end with this word or char
```

```python
data.filter(regex = "ion$").columns  #select column names that end with this word or char
```

```python
data.filter(regex = "mov+", axis = 1).columns  #select column names that end with this word or char
```

```python
data.filter(regex = "mov+", axis = 0).columns  #select column names that end with this word or char
```
